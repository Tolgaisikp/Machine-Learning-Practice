```python
import seaborn as sns
import numpy as np
import pandas as pd
df = sns.load_dataset("tips")
df.head()
```

```python
# 0-1 DÖNÜŞÜMÜ
```

```python
#Değişkenleri dönüştürürken mevcut bilginin yapısını bozuyormuyuz bozmuyormuyuz
```

```python
from sklearn.preprocessing import LabelEncoder
lbe = LabelEncoder()
```

```python
lbe.fit_transform(df["sex"])
```

```python
df["yeni_sex"] = lbe.fit_transform(df["sex"]) #eski halide kalsınki veri kayıpları yada daha sonradan lazım olursa diye
df.head()
```

```python
#ilgi odağımız genelde 1 olmak 
```

```python
#1 ve Diğerleri (0) dönüşümü
```

```python
df["yeni_day"] = np.where(df["day"].str.contains("Sun"),1,0)#sun değerleri 1 diğerleri 0 olsun
```

```python
df.head()
```

```python
#Çok Sınıflı Dönüşüm
```

```python
lbe.fit_transform(df["day"])
```

```python
df["yenii_day"] = lbe.fit_transform(df["day"])
df.head()
```

```python
#çok dikkat kategorik değişkenlerin birden fazla sınıfı varsa
```

```python
#ONE-HOT DÖNÜŞÜMÜ
```

```python
df.head()
```

```python
df_one_hot = pd.get_dummies(df, columns = ["sex"], prefix = ["sex"])
```

```python
df_one_hot.head()#erkekse sex male 1 kadınsa sex male 0
```
