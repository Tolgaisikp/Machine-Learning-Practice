```python
import numpy as np 
import pandas as pd
import seaborn as sns
```

```python
V1 = pd.array([1,3,6,5,7])
V2 = pd.array([7,7,5,8,12])
V3 = pd.array([6,12,5,6,14])

df = pd.DataFrame({"V1":V1,
                   "V2":V2,
                   "V3":V3})
df = df.astype(float)
df
```

```python
from sklearn import preprocessing
```

```python
preprocessing.scale(df)# ortalaması 0 standart sapması 1 işlemi
```

```python
#Normalizasyon
```

```python
preprocessing.normalize(df)
```

```python
#Min Max Dönüşümü
```

```python
scaler = preprocessing.MinMaxScaler(feature_range = (10,20))
scaler.fit_transform(df)#Varyans yapısı korundu ölçekleri değişti(öncekilerde dahil)
```

```python
#Binarize Dönüşüm
```

```python
#Standartlaştırma değiş Dönüştürmedir
```

```python
binarizer = preprocessing.Binarizer(threshold = 5).fit(df)
```

```python
binarizer.transform(df)#eşik değeri 5 verdiğimiz için altı ve üstü değerler için binary değer verir
```

```python
#0-1 Dönüşümü
```

```python
tips = sns.load_dataset("tips")
df = tips.copy()
df_l = df.copy()
```

```python
df_l.head()
```

```python
df_l["yeni_sex"] = df["sex"].cat.codes
```

```python
df_l.head()#cat ile ilk değer 1 olur preprocessing ile 0 olur
```

```python
# 1 ve diğerleri(0)
```

```python
df_l["yeni_day"] = np.where(df_l["day"].str.contains("Sun"),1,0)
```

```python
df_l.head(20)
```

```python
#Çok Sınıflı Dönüşüm
```

```python
lbe = preprocessing.LabelEncoder()
```

```python
df_l["daha_yeni_gun"] = lbe.fit_transform(df_l["day"])
```
